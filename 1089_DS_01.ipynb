```python
#7)Find Measure of Variance: Minimum, Maximum, Range, Mean Deviation, Standard Deviation, Co-efficient of Variation for Degree, Inter and 10th 
en.describe()
```

```python
#Range 
print("RANGE:")
print(" ")
print("degree:",en ['DEGREE'].max()-en ['DEGREE'].min())
print("intermediate:",en ['INTERMEDIATE'].max()-en ['INTERMEDIATE'].min())
print("SSC:",en ['SSC'].max()-en ['SSC'].min())
```

```python
#Mean Deviation
print("MEAN DEVIATION:")
print(" ")
print("degree:",en['DEGREE'].mad()) 
print("intermediate:",en['INTERMEDIATE'].mad())
print("SSC:",en ['SSC'].mad())
```

```python
#Co-efficient of Variation
print("Co-efficient of variation:")
print(" ")
print("degree:",en['DEGREE'].std()/en['DEGREE'].mean())
print("intermediate:",en['INTERMEDIATE'].std()/en['INTERMEDIATE'].mean())
print("SSC:",en['SSC'].std()/en['SSC'].mean())
```

```python
#8)Measures of Position: Standard Scores, Inter-quartile Range for Degree, Inter and 10th 
#Standard Score
print("Standard Score:")
print(" ")
print("degree:",stats.zscore(en['DEGREE']))
print("intermediate:",stats.zscore(en['INTERMEDIATE']))
print("SSC:",stats.zscore(en['SSC']))
```

```python
#Inter-quartile Range
print("Inter-quartile Range:")
print(" ")
q3,q1=np.percentile(en['DEGREE'],[7.5,2.5])
iqr=q3-q1
print("degree:",iqr)
q3,q1=np.percentile(en['INTERMEDIATE'],[75,25])
iqr=q3-q1
print("intermediate:",iqr)
q3,q1=np.percentile(en['SSC'],[75,25])
iqr=q3-q1
print("SSC:",iqr)
```

```python
#9)Create Box Plot and Identify Outliers for Degree, Inter and 10th 
boxplot = en.boxplot(column=['DEGREE', 'INTERMEDIATE', 'SSC'],rot=0, fontsize=12)  
```

```python
#outliers for Degree
print("outliers for Degree:")
print(" ")
q3,q1=np.percentile(en['DEGREE'],[7.5,2.5])
iqr=q3-q1
ul = q3+1.5*iqr
ll = q1-1.5*iqr
outliers = en['DEGREE'][(en['DEGREE'] > ul) | (en['DEGREE'] < ll)]
print("degree:",outliers)
#Outliers for Inter
q3,q1=np.percentile(en['INTERMEDIATE'],[75,25])
iqr=q3-q1
ul = q3+1.5*iqr
ll = q1-1.5*iqr
outliers = en['INTERMEDIATE'][(en['INTERMEDIATE'] > ul) | (en['INTERMEDIATE'] < ll)]
print("intermediate:",outliers)
#Outliers for 10th
q3,q1=np.percentile(en['SSC'],[75,25])
iqr=q3-q1
ul = q3+1.5*iqr
ll = q1-1.5*iqr
outliers = en['SSC'][(en['SSC'] > ul) | (en['SSC'] < ll)]
print("SSC:",outliers)
```
